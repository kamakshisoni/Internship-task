### Question: Explain dimension reduction in machine learning.
- 
- Dimensionality reduction is a process of reducing number of features from the data for better prediction by obtaining principal variables.
- there are two components of dimensionality reduction:
- feature selection: here we try to obtain a subset from the original data. that should contain only important variables that are essential for the prediction purpose. we can do this by applying filter,wrapper and embedded methods.
- feature extraction: this will reduce the high dimensional space data into low dimensional space.

- We have various methods to reduce dimensions such as PCA, LDA and GDA

### Question: How can you handle duplicate values in a dataset for a variable in Python using pandas?
-
- first we will import pandas library and read the data


```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\kamakshi\Desktop\New folder\new_file.csv")
data.head()
```

- As shown in above and below cell, there are duplicate values in the first column that is Name column and that should be unique.

```python
data.sort_values('first_name')
```

- with the help of pandas duplicate() method we can check for the duplicate values in the data. it will return bool value which mean False- unique value and True- duplicate value

```python
data.duplicated('first_name')[0:20]
```

```python
series = data['first_name'].duplicated()
series
```

#### this is the dataset including duplicate values

```python
data[series]
```

#### Now we will remove all those duplicate value from the first_name column

- In duplicate() method keep default value is True, so to remove duplicate value we will set it to False
- then save it to the variable - series2 
- it will keep the first value as unique and remove the rest of all

```python
series2 = data['first_name'].duplicated(keep=False)
df = data[~series2]
df
```

```python
df.first_name.duplicated()
```
